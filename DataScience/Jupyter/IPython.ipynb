## GLOBAL 変数(履歴)  
  * _i: 一個前の入力
  * _ii: 二個前の入力
  * _iii: 三個前の入力
  * _ih : 入力のリスト(input history)
  * _i<number\>: number番目の入力
  * _: 一個前の出力
  * __: 二個前の出力
  * ___: 三個前の出力
  * _oh: 出力履歴の辞書(output history)
  * _<number\>: number番目の出力
  * _dh: アクセスしたディレクトリの履歴(directory history)
