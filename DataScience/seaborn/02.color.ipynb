```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(rc={"figure.figsize": (8, 4)})
```

## パレット
* `color_palette`は`matplotlib`が支持する全ての色をパラメータとして指定できる
* `color_palette`を設定しないならデフォルトの色で絵を描く
* `set_palette`はあらゆるの図の色を設定する

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
sns.palplot(sns.color_palette('hls',12))
```

```python
data=np.random.normal(size=(20,12))+np.arange(12)/2
sns.boxplot(data=data,palette=sns.color_palette('hls', 12))
```

hls_palette()関数のパラメータで色を変えれる

* h: hue
* l: lightness
* s: saturation

```python
sns.palplot(sns.hls_palette(12,h=0.6, l=0.7,s=0.8))
```

```python
sns.palplot(sns.color_palette('Paired',12))
```
